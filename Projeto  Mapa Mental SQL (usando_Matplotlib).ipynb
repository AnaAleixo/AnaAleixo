```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.Graph()

# Adicionando os nós (comandos SQL e conceitos relacionados)
G.add_nodes_from(['SQL', 'SELECT', 'UPDATE', 'DELETE', 'INSERT INTO', 'CREATE DATABASE', 'ALTER DATABASE', 'CREATE TABLE', 'ALTER TABLE', 'DROP TABLE', 'CREATE INDEX', 'DROP INDEX', 'Banco de Dados', 'Tabela', 'Coluna', 'Índice'])

# Adicionando as arestas (relações entre os nós)
G.add_edges_from([('SQL', 'SELECT'), ('SQL', 'UPDATE'), ('SQL', 'DELETE'), ('SQL', 'INSERT INTO'), ('SQL', 'CREATE DATABASE'), ('SQL', 'ALTER DATABASE'), ('SQL', 'CREATE TABLE'), ('SQL', 'ALTER TABLE'), ('SQL', 'DROP TABLE'), ('SQL', 'CREATE INDEX'), ('SQL', 'DROP INDEX'), ('Banco de Dados', 'CREATE DATABASE'), ('Banco de Dados', 'ALTER DATABASE'), ('Banco de Dados', 'Tabela'), ('Tabela', 'CREATE TABLE'), ('Tabela', 'ALTER TABLE'), ('Tabela', 'DROP TABLE'), ('Tabela', 'Coluna'), ('Tabela', 'Índice'), ('Índice', 'CREATE INDEX'), ('Índice', 'DROP INDEX')])

# Definindo a posição dos nós no mapa mental
pos = nx.spring_layout(G)

# Desenhando o mapa mental
nx.draw_networkx_nodes(G, pos, node_size=2000)
nx.draw_networkx_labels(G, pos, font_size=12)
nx.draw_networkx_edges(G, pos, width=1)

# Exibindo o mapa mental
plt.axis('off')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Criando a figura em 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definindo a posição dos nós no espaço 3D
pos = {'SQL': (0, 0, 0), 'SELECT': (0, 1, 0), 'UPDATE': (0, 2, 0), 'DELETE': (0, 3, 0), 'INSERT INTO': (1, 0, 0), 'CREATE DATABASE': (1, 1, 0), 'ALTER DATABASE': (1, 2, 0), 'CREATE TABLE': (1, 3, 0), 'ALTER TABLE': (2, 0, 0), 'DROP TABLE': (2, 1, 0), 'CREATE INDEX': (2, 2, 0), 'DROP INDEX': (2, 3, 0), 'Banco de Dados': (0, 0, 1), 'Tabela': (1, 0, 1), 'Coluna': (2, 0, 1), 'Índice': (3, 0, 1)}

# Adicionando os nós ao gráfico
for node in pos:
    ax.scatter(pos[node][0], pos[node][1], pos[node][2], color='b')
    ax.text(pos[node][0], pos[node][1], pos[node][2], node, fontsize=10)

# Adicionando as conexões (arestas) entre os nós
connections = [('SQL', 'SELECT'), ('SQL', 'UPDATE'), ('SQL', 'DELETE'), ('SQL', 'INSERT INTO'), ('SQL', 'CREATE DATABASE'), ('SQL', 'ALTER DATABASE'), ('SQL', 'CREATE TABLE'), ('SQL', 'ALTER TABLE'), ('SQL', 'DROP TABLE'), ('SQL', 'CREATE INDEX'), ('SQL', 'DROP INDEX'), ('Banco de Dados', 'CREATE DATABASE'), ('Banco de Dados', 'ALTER DATABASE'), ('Banco de Dados', 'Tabela'), ('Tabela', 'CREATE TABLE'), ('Tabela', 'ALTER TABLE'), ('Tabela', 'DROP TABLE'), ('Tabela', 'Coluna'), ('Tabela', 'Índice'), ('Índice', 'CREATE INDEX'), ('Índice', 'DROP INDEX')]

for connection in connections:
    start = connection[0]
    end = connection[1]
    ax.plot([pos[start][0], pos[end][0]], [pos[start][1], pos[end][1]], [pos[start][2], pos[end][2]], 'k-')

# Definindo o nome dos eixos
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Exibindo o mapa mental em 3D
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.Graph()

# Adicionando os nós (comandos SQL e conceitos relacionados)
G.add_nodes_from(['SQL', 'SELECT', 'UPDATE', 'DELETE', 'INSERT INTO', 'CREATE DATABASE', 'ALTER DATABASE', 'CREATE TABLE', 'ALTER TABLE', 'DROP TABLE', 'CREATE INDEX', 'DROP INDEX', 'Banco de Dados', 'Tabela', 'Coluna', 'Índice'])

# Definindo as cores dos nós
node_colors = ['#FF6F61', '#5BC0EB', '#FDE74C', '#9BC53D', '#FFC48C', '#FC4445', '#FFE66D', '#A3E3ED', '#F7FFF7', '#D8E2DC', '#E2C2B9', '#8DC1C3', '#D9D9D9', '#D9D9D9', '#D9D9D9', '#D9D9D9']

# Adicionando as arestas (relações entre os nós)
G.add_edges_from([('SQL', 'SELECT'), ('SQL', 'UPDATE'), ('SQL', 'DELETE'), ('SQL', 'INSERT INTO'), ('SQL', 'CREATE DATABASE'), ('SQL', 'ALTER DATABASE'), ('SQL', 'CREATE TABLE'), ('SQL', 'ALTER TABLE'), ('SQL', 'DROP TABLE'), ('SQL', 'CREATE INDEX'), ('SQL', 'DROP INDEX'), ('Banco de Dados', 'CREATE DATABASE'), ('Banco de Dados', 'ALTER DATABASE'), ('Banco de Dados', 'Tabela'), ('Tabela', 'CREATE TABLE'), ('Tabela', 'ALTER TABLE'), ('Tabela', 'DROP TABLE'), ('Tabela', 'Coluna'), ('Tabela', 'Índice'), ('Índice', 'CREATE INDEX'), ('Índice', 'DROP INDEX')])

# Definindo a posição dos nós no mapa mental
pos = nx.spring_layout(G, k=0.4, iterations=50)

# Desenhando o mapa mental
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color=node_colors)
nx.draw_networkx_labels(G, pos, font_size=12)
nx.draw_networkx_edges(G, pos, width=1)

# Exibindo o mapa mental
plt.axis('off')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.DiGraph()

# Adicionando os nós (comandos SQL e conceitos relacionados)
G.add_nodes_from(['SQL', 'Banco de Dados', 'Tabela', 'Coluna', 'Índice', 'SELECT', 'UPDATE', 'DELETE', 'INSERT INTO', 'CREATE DATABASE', 'ALTER DATABASE', 'CREATE TABLE', 'ALTER TABLE', 'DROP TABLE', 'CREATE INDEX', 'DROP INDEX'])

# Adicionando as arestas (relações entre os nós)
G.add_edges_from([('SQL', 'Banco de Dados'), ('Banco de Dados', 'Tabela'), ('Tabela', 'Coluna'), ('Tabela', 'Índice'), ('SQL', 'SELECT'), ('SQL', 'UPDATE'), ('SQL', 'DELETE'), ('SQL', 'INSERT INTO'), ('Banco de Dados', 'CREATE DATABASE'), ('Banco de Dados', 'ALTER DATABASE'), ('Tabela', 'CREATE TABLE'), ('Tabela', 'ALTER TABLE'), ('Tabela', 'DROP TABLE'), ('Índice', 'CREATE INDEX'), ('Índice', 'DROP INDEX')])

# Definindo a posição dos nós no mapa mental
pos = {
    'SQL': (0, 1),
    'Banco de Dados': (1, 0),
    'Tabela': (2, -1),
    'Coluna': (3, -2),
    'Índice': (3, 0),
    'SELECT': (0, 2),
    'UPDATE': (0, 0),
    'DELETE': (0, -2),
    'INSERT INTO': (0, -4),
    'CREATE DATABASE': (1, -2),
    'ALTER DATABASE': (1, -4),
    'CREATE TABLE': (2, -2),
    'ALTER TABLE': (2, -4),
    'DROP TABLE': (2, -6),
    'CREATE INDEX': (3, -1),
    'DROP INDEX': (3, -3)
}

# Definindo as cores dos nós
colors = {
    'SQL': 'orange',
    'Banco de Dados': 'orange',
    'Tabela': 'orange',
    'Coluna': 'orange',
    'Índice': 'orange',
    'SELECT': 'green',
    'UPDATE': 'blue',
    'DELETE': 'red',
    'INSERT INTO': 'purple',
    'CREATE DATABASE': 'orange',
    'ALTER DATABASE': 'orange',
    'CREATE TABLE': 'orange',
    'ALTER TABLE': 'orange',
    'DROP TABLE': 'orange',
    'CREATE INDEX': 'orange',
    'DROP INDEX': 'orange'
}

# Desenhando o mapa mental
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color=[colors[n] for n in G.nodes()])
nx.draw_networkx_labels(G, pos, font_size=12)
nx.draw_networkx_edges(G, pos, width=1, arrowsize=20)

# Exibindo o mapa mental
plt.axis('off')
plt.show()
```

# **Alguns dos comandos SQL mais importantes<a>**

* SELECT- extrai dados de um banco de dados
* UPDATE- atualiza dados em um banco de dados
* DELETE- exclui dados de um banco de dados
* INSERT INTO- insere novos dados em um banco de dados
* CREATE DATABASE- cria um novo banco de dados
* ALTER DATABASE- modifica um banco de dados
* CREATE TABLE- cria uma nova tabela
* ALTER TABLE- modifica uma tabela
* DROP TABLE- exclui uma tabela
* CREATE INDEX- cria um índice (chave de pesquisa)
* DROP INDEX- exclui um índice
